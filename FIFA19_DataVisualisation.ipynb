```python
def strtofloat(x):
    try:
        if x[-1]=='M':
            return float(x[1:-1])*1000000
        else:
            return float(x[1:-1])*1000
    except:
        return np.nan
```

```python
df["Release Clause"]=df["Release Clause"].apply(strtofloat)
df['Release Clause']
```
