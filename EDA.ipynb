```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create():
    x=np.arange(10)
    y=x**2
    plt.plot(x,y)
    plt.title("line plot of square values")
    plt.xlabel("X-Axis")
    plt.ylabel("Y-Axis")
    plt.show()
```

```python
create()
```

```python
def sin():
    x=np.arange(0,10,0.1)
    y=np.sin(x)
    plt.plot(x,y,color="green")
    plt.title("Sine wave")
    plt.xlabel("X-AXIS")
    plt.ylabel("Y-AXIS")
    plt.xticks(rotation=69)
sin()
```
